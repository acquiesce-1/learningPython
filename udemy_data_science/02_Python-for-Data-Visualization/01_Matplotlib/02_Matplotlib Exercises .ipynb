```python
#####################################################################################
# Examples illustrating the use of plt.subplots().

# This function creates a figure and a grid of subplots with a single call, while
# providing reasonable control over how the individual plots are created.  For
# very refined tuning of subplot creation, you can still use add_subplot()
# directly on a new figure.
#####################################################################################
import matplotlib.pyplot as plt
import numpy as np

# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
plt.close('all')
```

```python
# Just a figure and one subplot
f, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
# Two subplots, the axes array is 1-d
f, ax = plt.subplots(nrows=2, sharex=True) #sharing xaxis
ax[0].plot(x, y)
ax[0].set_title('Sharing X axis')
ax[1].scatter(x, y)
```

```python
# Three subplots sharing both x/y axes
f, (ax1, ax2, ax3) = plt.subplots(nrows=3, sharex=True, sharey=True) #sharing both x/y axes

# Note that alternatively one could write ax[0], ax[1], etc...
ax1.plot(x, y)
ax1.set_title('Sharing both axes')

ax2.scatter(x, y)

ax3.scatter(x, 2 * y ** 2 - 1, color='r')

# Fine-tune figure; make subplots close to each other and hide x ticks for
# all but bottom plot.
f.subplots_adjust(hspace=0)
plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False)
```

```python
# row and column sharing
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex='col', sharey='row')

ax1.plot(x, y)
ax1.set_title('Sharing x per column, y per row')

ax2.scatter(x, y)

ax3.scatter(x, 2 * y ** 2 - 1, color='r')

ax4.plot(x, 2 * y ** 2 - 1, color='r')

f.subplots_adjust(hspace=0.2, wspace=0.05)
```

```python
# Four axes, returned as a 2-d array
f, ax = plt.subplots(nrows=2, ncols=2)
ax[0, 0].plot(x, y)
ax[0, 0].set_title('Axis [0,0]')

ax[0, 1].scatter(x, y)
ax[0, 1].set_title('Axis [0,1]')

ax[1, 0].plot(x, y ** 2)
ax[1, 0].set_title('Axis [1,0]')

ax[1, 1].scatter(x, y ** 2)
ax[1, 1].set_title('Axis [1,1]')

# Fine-tune figure; hide x ticks for top plots and y ticks for right plots
plt.setp([a.get_xticklabels() for a in ax[0, :]], visible=False)
plt.setp([a.get_yticklabels() for a in ax[:, 1]], visible=False)

plt.tight_layout()
```

```python
# Four polar axes
f, ax = plt.subplots(nrows=2, ncols=2, subplot_kw=dict(projection='polar'), figsize=(12,5))
ax[0, 0].plot(x, y)
ax[0, 0].set_title('Axis [0,0]')

ax[0, 1].scatter(x, y)
ax[0, 1].set_title('Axis [0,1]')

ax[1, 0].plot(x, y ** 2)
ax[1, 0].set_title('Axis [1,0]')

ax[1, 1].scatter(x, y ** 2)
ax[1, 1].set_title('Axis [1,1]')


# Fine-tune figure; make subplots farther from each other.
f.subplots_adjust(hspace=0.7, wspace=0.1)
plt.tight_layout()
```
