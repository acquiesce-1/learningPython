```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

#### Lets create data
x = np.linspace(0, 10, 20)
y = x ** 2


### Lets plot in object oriented mode
# Lets create an empty canvas herei refer to as 'figure' or 'fig'
fig = plt.figure()

# Now we need to give fig some positioning dimensions
axes = fig.add_axes([0, 0, 1, 1]) # % from left, bottom, width, height

# Now we plot
axes.plot(x,y,'r--')
axes.set_title('MY CHART')
axes.set_xlabel('My x Label')
axes.set_ylabel('My Y Label')
```

```python
##/// Plotting separate subplots

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,3), dpi=100)

# --- 1st plot
axes[0].set_title('PLot #1')
axes[0].plot(x,y,
             color='red',
             alpha=1,
             lw=1,
             ls='-',
             marker='o',
             markersize=1,
             markerfacecolor='red',
             markeredgewidth=2,
             markeredgecolor='red')


# --- 2nd Plot
axes[1].set_title('PLot #2')
axes[1].plot(y,x,
             color='blue',
             alpha=0.5,
             lw=1,
             ls='--',
             marker='+',
             markersize=7,
             markerfacecolor='blue',
             markeredgewidth=1,
             markeredgecolor='blue')

# --- 3rd Plot
axes[2].set_title('PLot #3')
axes[2].plot(x+2,y**3,
             color='green',
             alpha=0.2,
             lw=1,
             ls=':',
             marker='x',
             markersize=5,
             markerfacecolor='green',
             markeredgewidth=5,
             markeredgecolor='green')


# Lets cleanup the plots for aesthetics
plt.tight_layout()
```

```python
##/// Plotting on the same fig

fig, axes = plt.subplots(figsize=(12,3), dpi=100)

# --- 1st plot
axes.plot(x,y,
             color='red',
             alpha=1,
             lw=1,
             ls='-',
             marker='o',
             markersize=1,
             markerfacecolor='red',
             markeredgewidth=2,
             markeredgecolor='red')


# --- 2nd Plot
axes.plot(x+x,y+2,
             color='blue',
             alpha=0.5,
             lw=1,
             ls='--',
             marker='+',
             markersize=7,
             markerfacecolor='blue',
             markeredgewidth=1,
             markeredgecolor='blue')

# --- 3rd Plot
axes.plot(x*3,y*3,
             color='green',
             alpha=0.2,
             lw=1,
             ls=':',
             marker='x',
             markersize=5,
             markerfacecolor='green',
             markeredgewidth=5,
             markeredgecolor='green')


# Lets cleanup the plots for aesthetics
axes.set_title('MY PLOT')
plt.tight_layout()
```
