```python
import pandas as pd

# --- Read files in
filepath = r"C:\Users\danie\OneDrive\NON-PERSONAL\Education\Programming\dnPythonFiles\learningPython\udemy_data_science\02_Python-for-Data-Visualization"
filename = "Alinea_USA_windows.xlsx"
df = pd.read_excel(filepath + "\\" + filename, sheet_name = 'Sheet3')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df * 100

strategies = df.columns.tolist()
df.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

for strategy in strategies:
    df[strategy].plot(grid=True, figsize=(16,7), legend=strategy)

plt.title('Rolling 1 Year Return', fontsize=15, fontweight='bold')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
