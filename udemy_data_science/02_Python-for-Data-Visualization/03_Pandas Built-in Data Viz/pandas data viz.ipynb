```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
%matplotlib inline

df1 = pd.read_csv('df1', index_col=0) #where the index will equal the first column

df1.head()
```

```python
df2= pd.read_csv('df2')
df2.head()
```

```python
# Pandas has native ability to plot things

df1['A'].hist(bins=30)
```

```python
#Another way to call a plot is to use the .plot method
df1['A'].plot(kind='hist', bins=30)
```

```python
df2.plot.area()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.plot.line(x=df1.index, y='B', figsize=(12,3))
```

```python
df1.plot.scatter(x='A', y='B')
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*100)
```

```python
df1.plot.box()
```
