```python
from bs4 import BeautifulSoup
import urllib.request

tgt_url = "http://www.paulgraham.com/articles.html"
resp = urllib.request.urlopen(tgt_url)
soup = BeautifulSoup(resp, from_encoding=resp.info().get_param('charset'))

essay_link_list = []
root_url = "http://www.paulgraham.com/"
for link in soup.find_all('a', href=True):
    essay_link = root_url + link['href']
    essay_link_list.append(essay_link)
print("I am Done!")
```

```python
import urllib.request
from inscriptis import get_text

# Step 3: For url in urls:
    # urlib it and save to paul_graham_essays.txt (becareful not ot override prev text
count = 0
total_count = len(essay_link_list)
for essay_link in essay_link_list:
    print("Starting scrapping for: " + essay_link)
    count += 1
    url = essay_link
    html = urllib.request.urlopen(url).read().decode('ISO-8859-1')
    text = get_text(html)
    try:
        with open("paul_graham_essays.txt", "a+") as file:  
            file.write("\n" + essay_link + "\n" + text + "\n")
        print("Completed scrapping for: " + essay_link)
    except:
        print("Failed scrapping for: " + essay_link)
        pass
    print(round(int((count / total_count) * 100), 0))
    
# Step 4: Move text contents to Ms Word and make it look nicer
```
