```python
count = {}
for w in open('hub_doc_revs.txt').read().split():
    if w in count:
        count[w] += 1
    else:
        count[w] = 1
for word, times in count.items():
    print("%s was found %d times" % (word, times))
```

```python
words = []
mention = []

for word, times in count.items():
    words.append(word)
    mention.append(times)

```

```python
import pandas as pd

df = pd.DataFrame({'Word' : words,
                  'Times_Mentioned' :mention
                  })

df[df['Times_Mentioned'] > 10]
```
